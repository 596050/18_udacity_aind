## Cleaning

```python
import requests

# Fetch a web page
r = requests.get("https://www.udacity.com/courses/all")
print(r.text)
```

```python
import re

# Remove HTML tags using RegEx
pattern = re.compile(r'<.*?>')  # tags look like <...>
print(pattern.sub('', r.text))  # replace them with blank
```

```python
from bs4 import BeautifulSoup

# Remove HTML tags using Beautiful Soup library
soup = BeautifulSoup(r.text, "html5lib")
print(soup.get_text())
```

```python
# Find all course summaries
summaries = soup.find_all("div", class_="course-summary-card")
summaries[0]
```

```python
# Extract title
summaries[0].select_one("h3 a").get_text().strip()
```

```python
# Extract description
summaries[0].select_one("div[data-course-short-summary]").get_text().strip()
```

```python
# Find all course summaries, extract title and description
courses = []
summaries = soup.find_all("div", class_="course-summary-card")
for summary in summaries:
    title = summary.select_one("h3 a").get_text().strip()
    description = summary.select_one("div[data-course-short-summary]").get_text().strip()
    courses.append((title, description))

print(len(courses), "course summaries found. Sample:")
print(courses[0][0])
print(courses[0][1])
```

## Normalization

### Case Normalization

```python
# Sample text
text = "The first time you see The Second Renaissance it may look boring. Look at it at least twice and definitely watch part 2. It will change your view of the matrix. Are the human people the ones who started the war ? Is AI a bad thing ?"
print(text)
```

```python
# Convert to lowercase
text = text.lower() 
print(text)
```

### Punctuation Removal

```python
import re

# Remove punctuation characters
text = re.sub(r"[^a-zA-Z0-9]", " ", text) 
print(text)
```

## Tokenization

```python
# Split text into tokens (words)
words = text.split()
print(words)
```

### NLTK: Natural Language ToolKit

```python
# Another sample text
text = "Dr. Smith graduated from the University of Washington. He later started an analytics firm called Lux, which catered to enterprise customers."
print(text)
```

```python
from nltk.tokenize import word_tokenize

# Split text into words using NLTK
words = word_tokenize(text)
print(words)
```

```python
from nltk.tokenize import sent_tokenize

# Split text into sentences
sentences = sent_tokenize(text)
print(sentences)
```

```python
# List stop words
from nltk.corpus import stopwords
print(stopwords.words("english"))
```

```python
# Reset text
text = "The first time you see The Second Renaissance it may look boring. Look at it at least twice and definitely watch part 2. It will change your view of the matrix. Are the human people the ones who started the war ? Is AI a bad thing ?"

# Normalize it
text = re.sub(r"[^a-zA-Z0-9]", " ", text.lower())

# Tokenize it
words = text.split()
print(words)
```

```python
# Remove stop words
words = [w for w in words if w not in stopwords.words("english")]
print(words)
```

## Part-of-Speech Tagging

```python
from nltk import pos_tag

# Tag parts of speech (PoS)
sentence = word_tokenize("I always lie down to tell a lie.")
pos_tag(sentence)
```

### Sentence Parsing

```python
import nltk

# Define a custom grammar
my_grammar = nltk.CFG.fromstring("""
S -> NP VP
PP -> P NP
NP -> Det N | Det N PP | 'I'
VP -> V NP | VP PP
Det -> 'an' | 'my'
N -> 'elephant' | 'pajamas'
V -> 'shot'
P -> 'in'
""")
parser = nltk.ChartParser(my_grammar)

# Parse a sentence
sentence = word_tokenize("I shot an elephant in my pajamas")
for tree in parser.parse(sentence):
    print(tree)
```

```python
import IPython  # for displaying parse trees inline

# Visualize parse trees
for tree in parser.parse(sentence):
    tree.draw()
    #IPython.display.display(tree)  # instead of tree.draw()
```

```python
from nltk import pos_tag, ne_chunk
from nltk.tokenize import word_tokenize

# Recognize named entities in a tagged sentence
tree = ne_chunk(pos_tag(word_tokenize("Antonio joined Udacity Inc. in California.")))
#IPython.display.display(tree)
tree.draw()
```

## Stemming & Lemmatization

### Stemming

```python
from nltk.stem.porter import PorterStemmer

# Reduce words to their stems
stemmed = [PorterStemmer().stem(w) for w in words]
print(stemmed)
```

### Lemmatization

```python
from nltk.stem.wordnet import WordNetLemmatizer

# Reduce words to their root form
lemmed = [WordNetLemmatizer().lemmatize(w) for w in words]
print(lemmed)
```

```python
# Lemmatize verbs by specifying pos
lemmed = [WordNetLemmatizer().lemmatize(w, pos='v') for w in lemmed]
print(lemmed)
```
